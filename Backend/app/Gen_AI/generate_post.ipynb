```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate
from pydantic import BaseModel, Field
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
class Generate_Posts(BaseModel):
    # platform: str=Field(description="name of the platform")
    # post_title: str=Field(description="title of the post")
    # tone: str=Field(description="tone of the post")
    post_generated: str=Field(description="post generated")
```

```python
model=ChatGoogleGenerativeAI(model='gemini-flash-latest')
```

```python
parser=model.with_structured_output(Generate_Posts)
```

```python
prompt=PromptTemplate(
    template="Create a {platform} post on {post_title} having {tone} tone",
    input_variables=['platform','post_title','tone']
)
```

```python
chain=prompt|parser
```

```python
result=chain.invoke(
    {
        'platform': 'linkedin',
        'post_title': 'Layoffs due to AI',
        'tone': 'Serious'
    }
)
```
